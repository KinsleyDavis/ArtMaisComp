OpenCV (Open Source Computer Vision  http://opencv.org ) é essencialmente uma API C ++ . O pacote foca principalmente no processamento de imagem, captura e análise de vídeo, incluindo recursos como detecção de rosto e detecção de objeto.

> Bloco com recuo



```python
import cv2

def split_frames(filename, output_str):
    capture = cv2.VideoCapture(filename)

    i = 0
    while(capture.isOpened()):
        ret, frame = capture.read()
    
        if ret == False:
            break
        if i == 10:
            break
    
        cv2.imwrite(output_str.format(i),frame)
        i += 1

    capture.release()
```

Separa os frames em arquivos jpgs

```python
split_frames('video.mp4', 'kang-{}.jpg')
```
